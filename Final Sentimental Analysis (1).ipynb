# Percentage of Complex words

```python
text = article
```

```python
def calculate_percentage_complex_words(text):
    # tokenize text into words
    words = nltk.word_tokenize(text)
    num_words = len(words)
    
    # Calculate the number of complex words  with three or more syllables
    num_complex_words = sum(1 for word in words if syllable_count(word) >= 3)
    
    if num_words > 0:
        percentage_complex = (num_complex_words / num_words) * 100
    else:
        percentage_complex = 0
    
    return percentage_complex

def syllable_count(word):
    # Function to count syllables in a word
    return max(1, sum(word.lower().count(syllable) for syllable in ['a', 'e', 'i', 'o', 'u', 'y']))
```

```python
percentage_complex = calculate_percentage_complex_words(text)
```

# Word Count

```python
import string
```

```python
text2 = clean_article
```

```python
def count_cleaned_words(text2):
    # Tokenize text into words2
    words2 = word_tokenize(text2)
    
    # Remove punctuation from words2
    cleaned_words = [word.translate(str.maketrans('', '', string.punctuation)) for word in words2]
    
    # Remove empty strings after punctuation removal
    cleaned_words = [word for word in cleaned_words if word]
    
    # Count the cleaned words
    count = len(cleaned_words)
    
    return count
```

# Syallable Count Per Word

```python
# the syllables in each word ending with es or ed
```

```python
# let define text3 as the textual article tha we have defined it as text1
```

```python
def count_syllables(text1):
    # Count the number of vowels in the word
    vowels = 'aeiouy'
    num_vowels = sum(text1.lower().count(vowel) for vowel in vowels)
    
    # Handle exceptions like words ending with "es" or "ed"
    if text1.lower().endswith(('es', 'ed')) and num_vowels > 1:
        num_vowels -= 1
    
    # Ensure that at least one syllable is counted
    return max(1, num_vowels)

def syllable_count_per_word(text1):
    # Tokenize text into words
    words = word_tokenize(text1)
    
    # Count the number of syllables for each word
    syllable_counts = [count_syllables(word) for word in words]
    
    return syllable_counts
```

# Personal Pronouns

```python
import re
```

```python
def count_personal_pronouns(text1):
    # Define the list of personal pronouns
    personal_pronouns = ["I", "we", "my", "ours", "us"]

    # Define the regex pattern to match personal pronouns with word boundaries
    pattern = r'\b(?:' + '|'.join(personal_pronouns) + r')\b'

    # Find all matches of personal pronouns in the text (case-insensitive)
    matches = re.findall(pattern, text1, flags=re.IGNORECASE)

    # Exclude matches where "US" appears as a country name
    filtered_matches = [match for match in matches if match.lower() != 'us']

    # Count the occurrences of personal pronouns
    count = len(filtered_matches)

    return count
```

# Average Word Length

```python
def average_word_length(text1):
    # Tokenize text into words
    words = text1.split()
    
    # Calculate the total number of characters in each word
    total_characters = sum(len(word) for word in words)
    
    # Calculate the total number of words
    total_words = len(words)
    
    # Calculate the average word length
    if total_words > 0:
        average_length = total_characters / total_words
    else:
        average_length = 0
    
    return average_length
```
